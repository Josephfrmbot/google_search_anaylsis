# Import Librairies


```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib as plt
trends = TrendReq()
```

# Let's see if Kim Kardashian is really famous


```python
trends.build_payload(kw_list=["Kim Kardashian"])
data = trends.interest_by_region()
data = data.sort_values(by="Kim Kardashian", ascending=False)
data = data.head(15)

print(data)
```

# Where are the searches the most done ?

```python
data.reset_index().plot(x="geoName", 
                        y="Kim Kardashian", 
                        figsize=(15,12), kind="bar")
plt.style.use('fivethirtyeight')
plt.show()
```

# Total Searches for Kim Kardashian

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Kim Kardashian'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(15, 12))
data['Kim Kardashian'].plot()
plt.style.use('lol')
plt.title('Total Google Searches for Kim Kardashian', 
          fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```
